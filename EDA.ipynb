```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SDSS Datathon Cases/Real Estate/real-estate-data.csv")
```

```python
df.drop('id_', axis=1, inplace=True)
```

```python
df = df[df['price'].notna()]
```

```python
df['size'].value_counts()
```

```python
size_mapping = {
    '0-499 sqft': 0,
    '500-999 sqft': 1,
    '1000-1499 sqft': 2,
    '1500-1999 sqft': 3,
    '2000-2499 sqft': 4,
    '2500-2999 sqft': 5,
    '3000-3499 sqft': 6,
    '5500-3999 sqft': 7,  # Note: Verify this range; it seems out of expected order.
    '4000+ sqft': 8
}

# Map the 'size' column to numeric codes
df['size_encoded'] = df['size'].map(size_mapping)
```

```python
df
```

```python
df.drop('size', axis=1, inplace=True)
```

```python
df
```

```python
df['ward'].value_counts()
```

```python
# Create a mapping dictionary for exposure values
exposure_mapping = {
    'S': 0,
    'No': 1,
    'We': 2,
    'E': 3
}

# Map the exposure column to numeric codes
df['exposure_encoded'] = df['exposure'].map(exposure_mapping)
```

```python
df
```

```python
den_mapping = {
    'No': 0,
    'YES': 1
}

parking_mapping = {
    'N': 0,
    'Yes': 1
}

ward_mapping = {
    'W10': 0,
    'W11': 1,
    'W13': 2
}

# Apply the mappings to create new encoded columns
df['DEN_encoded'] = df['DEN'].map(den_mapping)
df['parking_encoded'] = df['parking'].map(parking_mapping)
df['ward_encoded'] = df['ward'].map(ward_mapping)
```

```python
df.drop(['DEN', 'parking', 'ward'], axis=1, inplace=True)
```

```python
df
```

```python
df.to_csv('cleaned_data.csv')
```
